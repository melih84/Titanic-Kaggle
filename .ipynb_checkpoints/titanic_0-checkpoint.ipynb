```python
import numpy as np
import pandas as pd
import seaborn as sns
import csv
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
```

```python
# let's read our dateset and look at it
data_train = pd.read_csv("train.csv") 
df_train = pd.DataFrame(data_train)
df_train.head()
```

```python
df_train['Age'].fillna(df_train['Age'].mean(),inplace=True) # Fills NaN values of ages with the mean of ages.
df_train['Age'] = df_train['Age'].round(decimals=0) # Rounds the ages to the nearest integer.
df_train['Embarked'].fillna('S',inplace=True) # Fills NaN values of Embarked with the most common value
df_train.head()
```

```python
df_train.hist(column='Survived',by='Sex',color='red') # Checks if sex and survival rate are correlated
```

```python
df_train.hist(column='Age',by='Survived',color='red') # Checks if age and survival rate are correlated
```

```python
df_train.hist(column='Survived',by='Pclass',color='red') # Checks if Pclass is correlated with survival rate
```

```python
df_train.hist(column='Survived',by='Embarked',color='red') # Checks if Embarked is correlated with survival rate
```

```python
df_train
```

```python
X_train = df_train[['Pclass','Sex','Age','SibSp','Parch','Embarked']]
```

```python
y_train = df_train['Survived']
```

```python
X_train
```

```python
X_train['Sex'].replace('female', 0,inplace=True)
X_train['Sex'].replace('male', 1,inplace=True)
X_train['Embarked'].replace('S', 0,inplace=True)
X_train['Embarked'].replace('C', 1,inplace=True)
X_train['Embarked'].replace('Q', 2,inplace=True)
```

```python
X_train
```

```python
#mean_age = X_train['Age'].mean()
#std_age = X_trainp['Age'].
#mean_age
x = X_train[['Age']].values.astype(float)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled)
X_train['Age'] = df_normalized
X_train['A']= 10
X_train
```

```python
X_train.isnull().sum() # Checs again if there is any nan value
```

```python
y_train
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_test
```

```python
# We have to do preprocessing to this data too
df_test['Age'].fillna(df_train['Age'].mean(),inplace=True) # Fills NaN values of ages with the mean of ages.
df_test['Age'] = df_train['Age'].round(decimals=0) # Rounds the ages to the nearest integer.
df_test['Embarked'].fillna('S',inplace=True) # Fills NaN values of Embarked with the most common value
X_test = df_test[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Embarked']]
X_test['Sex'].replace('female', 0,inplace=True)
X_test['Sex'].replace('male', 1,inplace=True)
X_test['Embarked'].replace('S', 0,inplace=True)
X_test['Embarked'].replace('C', 1,inplace=True)
X_test['Embarked'].replace('Q', 2,inplace=True)
x = X_test[['Age']].values.astype(float)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled)
X_test['Age'] = df_normalized

X_test

```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn import metrics
```



# K NEAREST NEIGHBORS 



```python
clf = KNeighborsClassifier(n_neighbors=3,n_jobs = 1)
clf.fit(X_train,y_train)
predictions = clf.predict(X_test)

submission = pd.DataFrame({
    'PassengerId': X_test['PassengerId'],
    'Survived': predictions
})
submission.to_csv('submission.csv', index=False)
submission = pd.read_csv('submission.csv')
submission.head()

```
